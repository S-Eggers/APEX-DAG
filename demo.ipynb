```python
from ApexDAG import DataFlowGraph

program = """
import polars as pl

df = pl.read_csv("test.csv")
df = df.with_columns(
    pl.col("a").cast(pl.Int32),
    pl.col("b").cast(pl.Float64),
    pl.col("c").cast(pl.Boolean),
)
df = df.get_dummies("a")
"""

dfg = DataFlowGraph()
dfg.parse_code(program)

G = dfg.get_graph()
G
```

```python
G.nodes(data=True)
```

```python
G.edges(data=True)
```

```python
from pathlib import Path
from logging import Logger
from ApexDAG.nn.training import GraphEncoder

encoder = GraphEncoder(Path("./"), Logger("test"), 1, 1, False)
encoder.encode_graphs([G], feature_to_encode="edge_type")
```
