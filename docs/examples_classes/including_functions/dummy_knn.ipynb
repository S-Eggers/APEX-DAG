```python
import numpy as np

import numpy as np

def tiny_knn_init(k=3):
    return {
        "k": k,
        "X": None,
        "y": None
    }

def tiny_knn_fit(model, X, y):
    model["X"] = np.array(X)
    model["y"] = np.array(y)

def tiny_knn_predict(model, Xq):
    Xq = np.array(Xq)
    preds = []
    for q in Xq:
        dist = ((model["X"] - q)**2).sum(axis=1)**0.5
        idx = dist.argsort()[:model["k"]]
        vals, cnts = np.unique(model["y"][idx], return_counts=True)
        preds.append(vals[cnts.argmax()])
    return preds


X = [[0,0],[1,1],[2,2],[8,8]]
y = [0,0,1,1]

clf = tiny_knn_init(k=1)

tiny_knn_fit(clf, X, y)
tiny_knn_predict(clf, [[1.5, 1.5], [7, 7]])
```
