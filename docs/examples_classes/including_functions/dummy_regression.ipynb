```python
import numpy as np

def tiny_linear_reg(lr=0.01, steps=200):
    state = {
        "lr": lr,
        "steps": steps,
        "w": 0.0,
        "b": 0.0
    }

    def fit(x, y):
        for _ in range(state["steps"]):
            pred = state["w"] * x + state["b"]
            dw = ((pred - y) * x).mean()
            db = (pred - y).mean()
            state["w"] -= state["lr"] * dw
            state["b"] -= state["lr"] * db

    def predict(x):
        return state["w"] * x + state["b"]

    return {
        "state": state,
        "fit": fit,
        "predict": predict
    }

x = np.array([1,2,3,4], float)
y = np.array([2,4,6,8], float)

m = tiny_linear_reg()
m["fit"](x, y)

print(m["predict"](np.array([5, 10])))
```
