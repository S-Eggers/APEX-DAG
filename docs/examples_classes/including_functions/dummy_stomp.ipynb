```python
import numpy as np

def tiny_stump():
    state = {"t": None}

    def fit(X, y):
        X_arr = np.array(X)
        y_arr = np.array(y)
        best_err, best_t = 1e9, None

        for t in np.unique(X_arr):
            pred = X_arr[:, 0] <= t
            err = (pred != y_arr).mean()
            if err < best_err:
                best_err, best_t = err, t

        state["t"] = best_t

    def predict(X):
        X_arr = np.array(X)
        return (X_arr[:, 0] <= state["t"]).astype(int)

    return {
        "state": state,
        "fit": fit,
        "predict": predict
    }

X = [[1],[2],[3],[10],[11]]
y = [0,0,0,1,1]

s = tiny_stump()
s["fit"](X, y)

print(s["predict"]([[2.5],[9]]))
```
