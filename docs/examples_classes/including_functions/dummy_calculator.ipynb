# Simple class notebook

Defines a small Calculator class and demonstrates its methods.

```python
calculator = {"value": 0}

def init(initial=0):
    calculator["value"] = initial
    return calculator["value"]

def add(x):
    calculator["value"] += x
    return calculator["value"]

def multiply(x):
    calculator["value"] *= x
    return calculator["value"]

def reset():
    calculator["value"] = 0
    return calculator["value"]
```

```python
init(5)
initial = calculator["value"]
after_add = add(3)
after_multiply = multiply(2)
after_reset = reset()
```
