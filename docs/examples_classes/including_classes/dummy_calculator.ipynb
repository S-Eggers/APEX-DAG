```python
class Calculator:
    """Simple calculator class."""
    def __init__(self, initial=0):
        self.value = initial

    def add(self, x):
        self.value += x
        return self.value

    def multiply(self, x):
        self.value *= x
        return self.value

    def reset(self):
        self.value = 0
        return self.value
```

```python
calc = Calculator(5)
initial = calc.value
after_add = calc.add(3)
after_multiply = calc.multiply(2)
after_reset = calc.reset()
```
