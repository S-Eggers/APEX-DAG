```python
import numpy as np

class TinyLinearReg:
    def __init__(self, lr=0.01, steps=200):
        self.lr, self.steps = lr, steps
        self.w, self.b = 0., 0.

    def fit(self, x, y):
        for _ in range(self.steps):
            pred = self.w * x + self.b
            dw = ((pred - y) * x).mean()
            db = (pred - y).mean()
            self.w -= self.lr * dw
            self.b -= self.lr * db

    def predict(self, x):
        return self.w * x + self.b

x = np.array([1,2,3,4], float)
y = np.array([2,4,6,8], float)
m = TinyLinearReg()
m.fit(x, y)
print(m.predict(np.array([5, 10]))) 
```
