```python
import numpy as np

class TinyKNN:
    def __init__(self, k=3):
        self.k = k

    def fit(self, X, y):
        self.X, self.y = np.array(X), np.array(y)

    def predict(self, Xq):
        preds = []
        for q in Xq:
            dist = ((self.X - q)**2).sum(axis=1)**0.5
            idx = dist.argsort()[:self.k]
            vals, cnts = np.unique(self.y[idx], return_counts=True)
            preds.append(vals[cnts.argmax()])
        return preds

X = [[0,0],[1,1],[2,2],[8,8]]
y = [0,0,1,1]
clf = TinyKNN(k=1)
clf.fit(X, y)
clf.predict([[1.5,1.5],[7,7]])
```
