```python
import numpy as np

class TinyStump:
    def fit(self, X, y):
        X, y = np.array(X), np.array(y)
        best_err, best_t = 1e9, None
        for t in np.unique(X):
            pred = X[:,0] <= t
            err = (pred != y).mean()
            if err < best_err:
                best_err, best_t = err, t
        self.t = best_t

    def predict(self, X):
        X = np.array(X)
        return (X[:,0] <= self.t).astype(int)

X = [[1],[2],[3],[10],[11]]
y = [0,0,0,1,1]
s = TinyStump()
s.fit(X, y)
print(s.predict([[2.5],[9]]))
```
